```python
import pandas as pd
from scipy import stats

# Đọc dữ liệu từ file CSV
df = pd.read_csv("shopeep_koreantop_clothing_shop_data.csv", encoding="UTF-8")
```

```python


# Kiểm định tương quan giữa rating_star và follower_count
df_rating_follower = df[['rating_star', 'follower_count']].dropna()
corr_coefficient, p_value = stats.pearsonr(df_rating_follower['rating_star'], df_rating_follower['follower_count'])
print("Tương quan giữa rating_star và follower_count:")
print("Hệ số tương quan:", corr_coefficient)
print("p-value:", p_value)
if p_value < 0.05:
    print("Có đủ cơ sở để bác bỏ giả thuyết không có tương quan.")
else:
    print("Không đủ cơ sở để bác bỏ giả thuyết không có tương quan.")

# Kiểm định tương quan giữa rating_star và item_count
df_rating_item = df[['rating_star', 'item_count']].dropna()
corr_coefficient, p_value = stats.pearsonr(df_rating_item['rating_star'], df_rating_item['item_count'])
print("\nTương quan giữa rating_star và item_count:")
print("Hệ số tương quan:", corr_coefficient)
print("p-value:", p_value)
if p_value < 0.05:
    print("Có đủ cơ sở để bác bỏ giả thuyết không có tương quan.")
else:
    print("Không đủ cơ sở để bác bỏ giả thuyết không có tương quan.")

# Kiểm định tương quan giữa is_shopee_verified và is_official_shop
df_verified_official = df[['is_shopee_verified', 'is_official_shop']].dropna()
contingency_table = pd.crosstab(df_verified_official['is_shopee_verified'], df_verified_official['is_official_shop'])
chi2, p, _, _ = stats.chi2_contingency(contingency_table)
print("\nTương quan giữa is_shopee_verified và is_official_shop:")
print("Giá trị Chi-squared:", chi2)
print("p-value:", p)
if p < 0.05:
    print("Có đủ cơ sở để bác bỏ giả thuyết không có tương quan.")
else:
    print("Không đủ cơ sở để bác bỏ giả thuyết không có tương quan.")
```
