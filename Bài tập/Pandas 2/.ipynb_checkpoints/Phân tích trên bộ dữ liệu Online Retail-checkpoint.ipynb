```python
import pandas as pd
df = pd.read_csv('C:\\Users\\ADMIN\\Downloads\\OnlineRetail.csv', encoding='latin1')
df.head()
```

```python
print('Số dòng của dữ liệu là: ', df.shape[0])
print('Số cột của dữ liệu là: ', df.shape[1])
```

```python
df.pivot_table(values = "Quantity", index = "Country", columns = "InvoiceNo", aggfunc = {'mean'})
```

```python
df.pivot_table(values = 'Quantity', index = 'CustomerID',columns = 'StockCode', aggfunc = {'min','max'})
```

```python
df.pivot_table(values = 'Quantity', index = 'StockCode', columns = 'Description', aggfunc = {'sum','max'})
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
pivot_by_date = df.pivot_table(values='Quantity', index=df['InvoiceDate'].dt.date, aggfunc='sum', fill_value=0)
print(pivot_by_date)
```
