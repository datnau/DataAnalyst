```python
# nhiều điều kiện
# df[...]
# lọc điều kiện:
  # điều kiện 1
  # điều kiện 2
  # điều kiện 3
  #...
# nối điều kiện:
  # &: và -> phải thỏa mãn cả 2 điều kiện
  # !: hoặc -> thỏa mãn 1 trong 2 điều kiện: hoặc điểm cao, hoặc đi học đầy đủ

# lấy dữ liệu thỏa mãn cả 2 điều kiện sau:
# join_year là năm 2021
# follower_count lớn hơn hoặc bằng 10000

condition1 = df['join_year'] == 2021
condition2 = df['follower_count'] >= 10000
final_condition = condition1 & condition2
result_df = df[final_condition]
print(result_df)
```
