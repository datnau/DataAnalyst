```python
#loc → gọi tên
#iloc → đếm
 
#. loc
#dòng trước cột sau

# Dòng
# Truy cập tới 1 dòng của dữ liệu: df. loc[tên địa chỉ dòng]
# Truy cập tới nhiều dòng liên tiếp của dữ liệu: df. loc[tên địa chỉ dòng a : tên địa chỉ dòng b]

# Cột
# Truy cập tới 1 cột cúa dữ liệu: df.locl:, tên cột]
# Truy cập tới nhiêu cột rời rạc của dữ liệu: df.loc[:, [tên cột A, tên cột B, tên cột CI]]
# Truy cập tới nhiều cột liên tiếp của dữ liệu: df. loc[:, tên cột A : tên cột E]

# Dòng và Cột
# Truy cập vào 1 giá trị có tên dòng và tên cột cụ thể: df. loc[tên dòng, tên cột]
# Truy cập tới nhiều dòng rời rạc và nhiều cột rời rạc của dữ liệu:
# df. loc[[tên dòng a, tên dòng b, tên dòng cl, [tên cột x, tên cột y, tên cột z]]
# Truy cập tới nhiều dòng liên tiếp và nhiều cột liên tiếp của dữ liệu:
# df. loc[tên dòng a : tên dòng d, tên cột x: tên cột z]
```
