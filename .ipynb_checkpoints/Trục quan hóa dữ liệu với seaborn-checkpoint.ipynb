```python
# Matplotlib
# plt.bar
# plt.plot
# plt.scatter
# plt.pie

# 1 biến
# Phân bố của dữ liệu - dữ liệu định lượng (numeric)
  #violinplot, histogram
# Đếm tần số cuả dữ liệu - dữ liệu định tính (categorical)
  # count plot

# 2 biến
# Mối quan hệ giữa 2 biến: scatterplot, lineplot
  # lmplot
# So sánh: range(phạm vi): box plot

# Nhiều hơn 2 biến:
# Mối quan giữa nhiều biến: heatmap

# Vẽ nhiều biểu đồ để thể hiện mối quan hệ giữa các biến: pairplot
```

```python
import matplotlib.pyplot as plt
import seaborn as sns #khai báo thư viện
import matplotlib.pyplot as plt
import seaborn as sns #khai báo thư viện
%matplotlib inline
sns.set(style='whitegrid', palette='hsv')

import pandas as pd
import numpy as np
```

```python
# chuẩn bị dữ liệu
# nhà hàng
tips = sns.load_dataset("tips")
tips.head()
```

```python
# tổng quan về nội dung dữ liệu
tips.describe()
```

```python
# Vẽ biểu đồ mô tả mối quan hệ giữa total_bill và tip
# scatter plot
plt.scatter(x=tips['total_bill'], y=tips['tip'])
```

```python
# hàm lmplot của seaborn để thể hiện mối quan hệ giữa 2 biến định lượng: total_bill, và tip
sns.lmplot(x='total_bill', y='tip', data=tips )
```

```python
# có sự khác biệt giữa người có hút thuốc và người không hút thuốc
sns.lmplot(x='total_bill',y= 'tip' , hue='smoker', data=tips)
```

```python
# có sự khác biệt giữa người có hút thuốc và người không hút thuốc
sns.lmplot(x='total_bill',y= 'tip' , col='smoker', data=tips)
```

```python
# tổng kết
# để biểu diễn mối quan hệ giữa 2 biến định lượng (numerical variable), có những cách sau
# cách 1: dùng matplotlib ->scatter plot
    # plt.scatter(x,y)
# cách 2: dùng lmplot của seaborn - scatterplot + lineplot
    # sns.lmplot(x,  y, data=)
    # Xem mối quan hệ giữa 2 biến định lượng
```

```python
# histogram
# phân bố của total_bill
tips['total_bill'].hist()
```

```python
# violinplot
sns.violinplot(y=tips['total_bill'])
```

```python
# so sánh phân bố biến định lượng theo nhóm
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
# so sánh phân bố biến định lượng theo nhóm
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
# xem phân bố/ range của biến định lượng
```

```python
# có sự khác nhau về phân bổ giữa nhóm hút thuốc và nhóm không hút thuốc
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips)
```

```python
# có sự khác nhau giữa giờ ăn hay không: time
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)
```

```python
# tổng kết
# phân bố của dữ liệu
# cách 1: dùng histogram để xem phân bổ của 1 biến định lượng: df['total_bill'].hist()
# cách 2: dùng violinplot

# phân bổ của 1 biến định lượng theo nhóm:
# ví dụ: xem phân bổ/ khoảng của total_bill theo các ngày trong tuần
# cách 1: dùng boxplot: sns.boxplot(x= , y= , data=)
# cách 2: dùng violinplot: sns.violinplot(x= ,y= ,data= )

# xem phân bổ của dữ liệu định tính
# đếm số lượng dữ liệu trong mỗi giá trị định tính
# sns.countplot(...)
```

```python
# giới tính: Nam, Nữ
# đếm trong tập dữ liệu có bao nhiêu Nam, bao nhiêu Nữ
# countplot()
sns.countplot(x='sex' , hue='smoker', data= tips)
```

```python
# vẽ phần chính
sns.scatterplot()
sns.lineplot()
sns.barplot()
sns.countplot()
sns.histplot()
sns.boxplot()
sns.heatmap()
sns.pairplot()
sns.lmplot()

# Vẽ trang trí -> matplotlib
# chỉ định kích cỡ của hình: plt.figure(figsize=())
# tên biểu đồ: plt.title('ten')
# tên trục x: plt.xlabel('ten truc x')
# tên trục y: plt.ylabel('ten truc y')
# Kết hợp biểu đồ: 
  # Cùng trục
  # Cùng khung

# mối quan hệ: 2 biến
# phân bố: 1 biến theo nhóm
#
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(8, 6))
sns.heatmap( tips[['tip', 'total_bill', 'size']].corr() , annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap between Tip, Total Bill, and Size')
plt.show()

```
